```python
import numpy as np
import pandas as pd
import tensorflow as tf
import autokeras as ak
```

```python
train_file_path = "train_data.csv"
test_file_path = "test_data.csv"
```

```python
# Initialize the structured data regressor.
reg = ak.StructuredDataRegressor(
    overwrite=True, max_trials=3 # 모델 개수 
) 
reg.fit(
    train_file_path,
    "OUT", # label
    epochs=100, 
)
```

```python
predicted_y = reg.predict(test_file_path)
submission=pd.read_csv('answer_sample.csv')
submission['OUT']=predicted_y
submission.to_csv('answer_sample1.csv', index=False)
```

```python
model = reg.export_model()

print(type(model))  # <class 'tensorflow.python.keras.engine.training.Model'>

try:
    model.save("model_autokeras", save_format="tf")
except Exception:
    model.save("model_autokeras.h5")


loaded_model = tf.keras.models.load_model("model_autokeras", custom_objects=ak.CUSTOM_OBJECTS)

#predicted_y = loaded_model.predict(test_file_path)
#predicted_y
```
